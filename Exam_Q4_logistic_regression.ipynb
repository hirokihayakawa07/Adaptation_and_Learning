## Adaptation and Learning Exam Q4

#### Initial setting

```python
# import libraries
import numpy as np
import numpy.linalg as lg
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
# set parameters
params = {
    'rho': 1.0,
    'mu':0.01,
    'M': 10,
    'N': 100,
    'maxit': 5000,
    'w_gt': np.random.randn(10),
    'w_init': np.random.randn(10),
    'alpha': 1.0,
    'beta1': 0.90,
    'beta2': 0.999,
    'epsilon': 10**-8
}
```

```python
# create data set
h = np.random.randn(params['M'],params['N'])
gamma = np.transpose(h)@params['w_gt']
```

#### Define required functions

```python
def Pw(w, gamma, h, params):
    return params['alpha']*lg.norm(w,1) + params['rho']*lg.norm(w) + 1/params['N']*lg.norm(gamma - np.transpose(h)@w)**2
```

```python
def gradient(w, gamma_m, h_m, params):
    return 2*params['rho']*w - 2*h_m*(gamma_m - np.transpose(h_m)@w)
```

```python
def l1_subgradient(w,params):
    return params['alpha']*np.sign(w)
```

```python
def l1_prox(z, params):
    mu = params['mu']
    alpha = params['alpha']
    z = np.where((z > -mu*alpha) & (z < mu*alpha), 0, z)
    z = np.where(z <= -mu*alpha, z+mu*alpha, z)
    z = np.where(z >= mu*alpha, z-mu*alpha, z)

    return z
```

```python
def subgradient(gamma, h, params):
    mu = params['mu']
    w = params['w_init']
    loss = []
    for i in range(params['maxit']):
        loss.append(Pw(w, gamma, h, params))
        j = np.random.randint(params['N'])
        w = w - mu*(gradient(w,gamma[j],h[:,j],params) + l1_subgradient(w,params))
    return loss
```

```python
loss_subgradient = subgradient(gamma, h, params)
```

```python
# check convergence with the algorithm
x = range(params['maxit'])
loss = loss_subgradient
y = loss
# y = abs(loss-loss[-1]) / (max(loss)-loss[-1])
print(loss[-1])
plt.plot(x,y)
plt.yscale('log')
plt.show()
```

```python
def prox_gradient(gamma, h, params):
    mu = params['mu']
    w = params['w_init']
    loss = []
    for i in range(params['maxit']):
        loss.append(Pw(w, gamma, h, params))
        j = np.random.randint(params['N'])
        z = w - mu*gradient(w,gamma[j],h[:,j],params)
        w = l1_prox(z,params)
    return loss
```

```python
loss_prox = prox_gradient(gamma, h, params)
```

```python
# check convergence with the algorithm
x = range(params['maxit'])
loss = loss_prox
y = loss
# y = abs(loss-loss[-1]) / (max(loss)-loss[-1])
print(loss[-1])
plt.plot(x,y)
plt.yscale('log')
plt.show()
```

```python
def adam(gamma, h, params):
#     np.random.seed(seed=1)
    beta1 = params['beta1']
    beta2 = params['beta2']
    mu = params['mu']
    epsilon = params['epsilon']
    b_bar = np.zeros(params['M'])
    s = np.zeros(params['M'])
    w = params['w_init']
    loss = []
    for i in range(params['maxit']):
        loss.append(Pw(w, gamma, h, params))
        j = np.random.randint(params['N'])
        b = gradient(w, gamma[j], h[:,j], params) + l1_subgradient(w,params)
        b_bar = beta1*b_bar + (1-beta1)*b
        s = beta2*s + (1-beta2)*(b**2)
        A = np.diag(np.sqrt(s)) + epsilon*np.eye(params['M'])
        w = w - mu*(np.sqrt(1 - beta2**(i+1)) / (1 - beta1**(i+1))) * lg.inv(A)@b_bar
    return loss
```

```python
loss_adam = adam(gamma,h,params)
```

```python
# check convergence with the algorithm
x = range(params['maxit'])
loss = loss_adam
y = loss
# y = abs(loss-loss[-1]) / (max(loss)-loss[-1])
print(loss[-1])
plt.plot(x,y)
plt.yscale('log')
plt.show()
```

```python
# normalization of the calculated loss values
loss = np.zeros((3,params['maxit']))
for i in range(10):
    loss[0,:] += np.array( abs(subgradient(gamma, h, params) - subgradient(gamma, h, params)[-1]) / (subgradient(gamma, h, params)[0] - subgradient(gamma, h, params)[-1]))/10
    loss[1,:] += np.array( abs(prox_gradient(gamma, h, params) - prox_gradient(gamma, h, params)[-1]) / (prox_gradient(gamma, h, params)[0] - prox_gradient(gamma, h, params)[-1]))/10
    loss[2,:] += np.array( abs(adam(gamma, h, params) - adam(gamma, h, params)[-1]) / (adam(gamma, h, params)[0] - adam(gamma, h, params)[-1]))/10
```

```python
# plot the normalized loss
fig = plt.figure(figsize = (10,10))
ax1 = plt.axes()

x = range(params['maxit'])
ax1.plot(x, loss[0])
ax1.plot(x, loss[1])
ax1.plot(x, loss[2])
ax1.set_xlabel("epock")
ax1.set_ylabel("P(wn) - P(w_star)")
ax1.set_yscale('log')
ax1.grid(True,which="both",ls="-")
ax1.legend(['Subgradient','Prox_gradient','ADAM'])
ax1.set_title('Learning curves (logarithmic scale)')

fig.tight_layout()
```
